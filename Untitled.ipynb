### Feature Reduction Results

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pylab
```

```python
childrensFiction = {100: 76.6,50: 76.6, 20: 76.4, 10: 76.7, 5: 76.6, 1: 76.4}
scienceFiction = {100: 64.5, 50: 64.5, 20: 63.6, 10: 62.3, 5: 61.5, 1: 60.2}
gothicFiction = {100: 57.6, 50: 57.8, 20: 57.0, 10: 57.2, 5: 57.7, 1: 58.2}
fig, ax = plt.subplots(figsize=(9, 7))
ax.plot(childrensFiction.keys(), childrensFiction.values(),'bs', label='childrens fiction', markersize=10)
ax.plot(scienceFiction.keys(), scienceFiction.values(), 'ro', label='science fiction', markersize=10)
ax.plot(gothicFiction.keys(), gothicFiction.values(), 'go', label='gothic fiction', markersize=10)
plt.xticks(scienceFiction.keys())
#invert the x axis 
plt.gca().invert_xaxis()
fig.suptitle('Feature Reduction for', fontsize=16)
plt.title('Children\'s Fiction, Science Fiction, and Gothic Fiction', fontsize=16)
plt.xlabel('Percentage of features used', fontsize=16)
plt.ylabel('Accuracy (%)', fontsize=16)
plt.legend()
pylab.ylim(50, 90)
plt.show()
```
